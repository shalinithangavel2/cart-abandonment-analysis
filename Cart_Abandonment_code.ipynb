```python
# Function to calculate risk score
def calculate_risk(row):
    score = 0
    
    # Purchase behavior
    if row["Purchase_Status"] == "No":
        score += 3
    elif row["Purchase_Status"] == "Sometimes":
        score += 2
    
    # Shopping frequency
    if row["Online_Shopping_Frequency"] == "Rarely":
        score += 2
    elif row["Online_Shopping_Frequency"] == "Sometimes":
        score += 1
    
    # Abandonment reason (price/delivery related)
    reason = str(row["Abandonment_Reason"]).lower()
    if "price" in reason or "delivery" in reason:
        score += 1
    
    return score
```

```python
df["Risk_Score"] = df.apply(calculate_risk, axis=1)
```

```python
df[["Purchase_Status", "Online_Shopping_Frequency", "Risk_Score"]].head()
```

```python
def risk_level(score):
    if score >= 4:
        return "High Risk"
    elif score >= 2:
        return "Medium Risk"
    else:
        return "Low Risk"

df["Risk_Level"] = df["Risk_Score"].apply(risk_level)
```
